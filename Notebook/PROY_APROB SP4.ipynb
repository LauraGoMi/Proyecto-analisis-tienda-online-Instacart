### [C3] Para cada producto, ¬øcu√°l es la proporci√≥n de las veces que se pide y que se vuelve a pedir?

````python
#Tasa de repetici√≥n = (N√∫mero de pedidos repetidos del producto) / (Total de pedidos del producto)
#Filtrar solo productos reordenados (reordered == 1)
productos_reordenados = df_order_products[df_order_products['reordered'] == 1]

#Contar cu√°ntas veces aparece cada product_id en los reordenados
top_reordenados = productos_reordenados['product_id'].value_counts().head(20)


total_pedidos = df_order_products.groupby('product_id').size()

# Filtrar primero, luego agrupar y contar
df_filtrado = df_order_products [df_order_products['reordered']== 1]
reordenes = df_filtrado.groupby('product_id').size()

# Calcular la proporci√≥n de reorden para cada producto
proporcion_reorden = reordenes / total_pedidos

# Llenar valores NaN con 0 (productos que nunca fueron reordenados)
proporcion_reorden = proporcion_reorden.fillna(0)

print("Primeras 10 proporciones:")
print(proporcion_reorden.head(10))
````

````python
# Obtener los 20 productos con mayor tasa de repetici√≥n
top_20_proporciones = proporcion_reorden.nlargest(20)

# Crear DataFrame para unir con nombres de productos
top_20_prop_df = pd.DataFrame({
    'product_id': top_20_proporciones.index,
    'proporcion_reorden': top_20_proporciones.values
})

# Unir con nombres de productos
top_20_prop_con_nombres = top_20_prop_df.merge(df_products, on='product_id', how='left')

print("TOP 20 PRODUCTOS CON MAYOR TASA DE REPETICI√ìN:")
print("=" * 60)
for i, row in top_20_prop_con_nombres.iterrows():
    print(f"{i+1:2d}. ID: {row['product_id']:5d} | "
          f"Tasa: {row['proporcion_reorden']:.3f} | "
          f"Nombre: {row['product_name']}")
 # histograma

# Crear histograma de la distribuci√≥n de tasas de repetici√≥n
plt.figure(figsize=(12, 6))
proporcion_reorden.hist(bins=50, alpha=0.7, color='skyblue', edgecolor='navy')
plt.title('Distribuci√≥n de las Tasas de Repetici√≥n por Producto', fontsize=14, fontweight='bold')
plt.xlabel('Tasa de Repetici√≥n (0 = nunca se repite, 1 = siempre se repite)')
plt.ylabel('N√∫mero de Productos')
plt.grid(True, alpha=0.3)

# Agregar estad√≠sticas en el gr√°fico
media = proporcion_reorden.mean()
mediana = proporcion_reorden.median()
plt.axvline(media, color='red', linestyle='--', label=f'Media: {media:.3f}')
plt.axvline(mediana, color='orange', linestyle='--', label=f'Mediana: {mediana:.3f}')
plt.legend()
plt.show()

# Mostrar estad√≠sticas descriptivas
#print("ESTAD√çSTICAS DE LAS TASAS DE REPETICI√ìN:")
#print("=" * 40)
#print(f"Media: {proporcion_reorden.mean():.3f}")
#print(f"Mediana: {proporcion_reorden.median():.3f}")
#print(f"Desviaci√≥n est√°ndar: {proporcion_reorden.std():.3f}")
#print(f"M√≠nimo: {proporcion_reorden.min():.3f}")
#print(f"M√°ximo: {proporcion_reorden.max():.3f}")
````

### [C4] Para cada cliente, ¬øqu√© proporci√≥n de sus productos ya los hab√≠a pedido?

````python
#Proporci√≥n = (Productos reordenados en el pedido) / (Total de productos en el pedido)

# Para cada pedido, calcular la proporci√≥n de productos reordenados
proporciones_por_pedido = df_order_products.groupby('order_id').agg({
    'reordered': ['sum', 'count']  # suma de reordered y total de productos
}).round(3)

# Aplanar las columnas
proporciones_por_pedido.columns = ['productos_reordenados', 'total_productos']

# Calcular la proporci√≥n
proporciones_por_pedido['proporcion_reorden'] = (
    proporciones_por_pedido['productos_reordenados'] / 
    proporciones_por_pedido['total_productos']
)

# Unir con informaci√≥n del cliente
proporciones_con_cliente = proporciones_por_pedido.merge(
    df_instacart_orders[['order_id', 'user_id']], 
    on='order_id', 
    how='left'
)

print("Primeras 10 proporciones por pedido:")
print(proporciones_con_cliente.head(10))

#Paso 2: Agrega este c√≥digo en la celda 119:
# Calcular estad√≠sticas por cliente
estadisticas_por_cliente = proporciones_con_cliente.groupby('user_id').agg({
    'proporcion_reorden': ['mean', 'count', 'std']
}).round(3)

# Aplanar columnas
estadisticas_por_cliente.columns = ['proporcion_promedio', 'num_pedidos', 'desviacion_std']

print("Estad√≠sticas generales:")
print(f"Proporci√≥n promedio general: {estadisticas_por_cliente['proporcion_promedio'].mean():.3f}")
print(f"Clientes con mayor lealtad (proporci√≥n > 0.8): {(estadisticas_por_cliente['proporcion_promedio'] > 0.8).sum():,}")
print(f"Clientes exploradores (proporci√≥n < 0.2): {(estadisticas_por_cliente['proporcion_promedio'] < 0.2).sum():,}")

# Mostrar distribuci√≥n
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
estadisticas_por_cliente['proporcion_promedio'].hist(bins=50, alpha=0.7, color='lightgreen', edgecolor='darkgreen')
plt.title('Distribuci√≥n de Proporciones de Reorden por Cliente')
plt.xlabel('Proporci√≥n Promedio de Productos Reordenados')
plt.ylabel('N√∫mero de Clientes')
plt.grid(True, alpha=0.3)
plt.show()

````
